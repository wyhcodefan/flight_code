```python
import pandas as pd
from datetime import date
import numpy as np
import matplotlib.pyplot as plt
import datetime

data = pd.read_csv('CA1883.csv')
data.columns = ['TICKET_DATE','FLIGHT_DATE','REMAIN_DAY','FLIGHT_NO','START_AIRPORT',    #重新命名列名
               'ARRIVAL_AIRPORT','CABIN','ValidFare']
data = data[data.ValidFare != '\\N']
data = data[['TICKET_DATE','FLIGHT_DATE','REMAIN_DAY','FLIGHT_NO','START_AIRPORT',
               'ARRIVAL_AIRPORT','ValidFare']]
data = data[data.REMAIN_DAY <=45]
data.ValidFare = data.ValidFare.astype('int')
temp = data.groupby(['REMAIN_DAY','FLIGHT_DATE'])['ValidFare'].agg('min')   #根据remain_day和flight_day进行分组，然后将得到的结果，在求各分组中validfare最小值
temp = temp.reset_index()     #给上一行得到的结果重新设置行标号
temp = temp[['REMAIN_DAY','FLIGHT_DATE','ValidFare']]
data = pd.merge(temp,data,how='left',on=['REMAIN_DAY','FLIGHT_DATE','ValidFare']).drop_duplicates()

start_date = '2017-07-21'
end_date  = '2018-09-01'
#验证集的起点和终点为2018年7月17日到7月23日这7天，后面的数据可以用来计算起飞前最低价，但是不出现在验证集的数据集中
validation_start = '2018-07-17'
validation_end = '2018-07-23'

train_set = data[data.TICKET_DATE < validation_start]

#先删除下异常点
min_price_threshold = np.percentile(np.array(train_set.ValidFare),0.1)+10 #0.1%分位数+10元
max_price_threshold = np.percentile(np.array(train_set.ValidFare),99.9)+10
train_set = train_set[(train_set.ValidFare>min_price_threshold)&(train_set.ValidFare<max_price_threshold)].reset_index(drop=True)
validation_set  = data[(data.TICKET_DATE >=validation_start)&(data.FLIGHT_DATE <= end_date)].reset_index(drop=True)
#print(len(train_set))

#train_set.to_csv('data_feature/train_data.csv',index=None)
#validation_set.to_csv('data_feature/validation_data.csv',index=None)
#data.to_csv('data_feature/all_data.csv',index=None)
    
```

```python
#计算未来最低价，并解决尖刺问题
#三个gap作为影响范围的阈值

def extra_feature1_1(data):
    
    Y_price = 1240
    gap1 = Y_price/13
    gap2 = Y_price/13*2
    gap3 = Y_price/13*3
    
    #fare_list_dict = create_dict
    
    data_result = copy.deepcopy(data)
    data_result = data_result.sort_values(by=['FLIGHT_DATE','REMAIN_DAY'],ascending=True)
    
    data_result['low_predict'] = 100000
    data_result['median_price_lower_than_now_price'] = -1
    data_result['mean_price_lower_than_now_price'  ] = -1
    data_result['lower_day_percent'] = 0
    data_result['avg_price'] = -1
    
    column_lens = len(data_result.columns)     #获取data_result的列名
    data_array = data_result.as_matrix()       #将DataFrame类型转换成array类型
    dateset_columns = data_result.columns
    
    LAST_FLIGHT_DATE = 'test'
    FARE_LIST = []
    
    for index_temp in range(len(data_array)):      #对data_array各条数据进行处理
        
        NOW_REMAIN_DAY  = data_array[index_temp][0]
        NOW_FLIGHT_DATE = data_array[index_temp][1]
        NOW_ValidFare   = data_array[index_temp][2]
        
        if(NOW_FLIGHT_DATE == LAST_FLIGHT_DATE):
            FARE_LIST.append(NOW_ValidFare)
        else:
            FARE_LIST = []
            FARE_LIST.append(NOW_ValidFare)
            LAST_FLIGHT_DATE = NOW_FLIGHT_DATE
            
        threshold_gap1 = NOW_ValidFare - gap1
        threshold_gap2 = NOW_ValidFare - gap2
        threshold_gap3 = NOW_ValidFare - gap3
        
        low_price_dict_gap1 = {}
        low_price_dict_gap2 = {}
        low_price_dict_gap3 = {}
        
        min_price_temp = 100000
        sum_price = 0
        
        for increment_index in range(len(FARE_LIST)-1):#遍历remainday小于当前remainday的机票价格，然后来给每一个low_price_dict_gap赋值
            ValidFare_temp = FARE_LIST[increment_index]
            if(int(ValidFare_temp) < NOW_ValidFare):
                if(int(ValidFare_temp)>=threshold_gap1):
                    low_price_dict_gap1[str(increment_index)] = int(ValidFare_temp)
                elif(int(ValidFare_temp)>=threshold_gap2):
                    low_price_dict_gap1[str(increment_index)] = int(ValidFare_temp)
                    low_price_dict_gap2[str(increment_index)] = int(ValidFare_temp)
                else:
                    low_price_dict_gap1[str(increment_index)] = int(ValidFare_temp)
                    low_price_dict_gap2[str(increment_index)] = int(ValidFare_temp)
                    low_price_dict_gap3[str(increment_index)] = int(ValidFare_temp)
            elif(int(ValidFare_temp)<min_price_temp):
                min_price_temp = int(ValidFare_temp)
                
        def analysis_low_price_dict(low_price_dict,influence_day_gap=5,REMAIN_DAY=NOW_REMAIN_DAY):
            
        #根据现在填满的低价字典来根据影响范围理论来计算出适合的最低价
            if(low_price_dict !={}):
                key_list = list(low_price_dict.keys())
                key_list.sort(reverse=True)
                influence_dict = {}
                for temp in range(len(key_list)):
                    power_temp = 1
                    influence_day_gap_temp = influence_day_gap**power_temp
                    for temp2 in range(temp+1,len(key_list)):
                        if(int(key_list[(temp2)]) >= (int(key_list[temp]) - influence_day_gap_temp)):
                            power_temp+=1
                            influence_day_gap_temp = influence_day_gap**power_temp
                        else:
                            break
                    influence_dict[key_list[temp]] = int(key_list[temp]) + influence_day_gap_temp
                if(influence_dict!={}):
                    if((np.max(list(influence_dict.values()))>=REMAIN_DAY)):
                        return np.percentile(list(low_price_dict.values()),50)#当前返回的是中位数，后面激进起来就可以返回更小的下四分位数之类的
            return None
        
        gap3_result = analysis_low_price_dict(low_price_dict_gap3)
        if(gap3_result):
            data_array[index_temp,column_lens-5] = gap3_result
        else:
            gap2_result = analysis_low_price_dict(low_price_dict_gap2)
            if(gap2_result):
                data_array[index_temp,column_lens-5] = gap2_result
            else:
                gap1_result = analysis_low_price_dict(low_price_dict_gap1)
                if(gap1_result):
                    data_array[index_temp,column_lens-5] = gap1_result
                else:
                    data_array[index_temp,column_lens-5] = min_price_temp
        if(len(low_price_dict_gap1)>0):
            data_array[index_temp,column_lens-4] = np.median(list(low_price_dict_gap1.values()))
            data_array[index_temp,column_lens-3] = np.mean(list(low_price_dict_gap1.values()))
        else:
            data_array[index_temp,column_lens-4] = NOW_ValidFare
            data_array[index_temp,column_lens-3] = NOW_ValidFare
       
        if(len(FARE_LIST)>1):
            data_array[index_temp,column_lens-2] = len(list(low_price_dict_gap1.values()))/NOW_REMAIN_DAY
            data_array[index_temp,column_lens-1] = np.mean(FARE_LIST[:-1])
        else:
            data_array[index_temp,column_lens-1]   = NOW_ValidFare
    
    return pd.DataFrame(data = data_array,columns = dateset_columns)
            
```

```python

def extra_feature2_1(data):
    
    data_result = copy.deepcopy(data)#一定要用deepcopy，不然在函数内对data加了列，会在函数外也有效果

    data_result = data_result.sort_values(by=['FLIGHT_DATE','REMAIN_DAY'],ascending=[True,False])
    data_result['ValidFare_is_most_low']  = -1
    data_result['ValidFare_minus_most_low']  = -1
    data_result['ValidFare_minus_yesterday']  = -1
    data_result['previous_high_price']  = -1
    data_result['previous_avg_price']  = -1
    data_result['previous_low_price']  = -1
    
    dateset_columns = data_result.columns
    columns_len  = len(dateset_columns)
    data_array = data_result.as_matrix()
    
    LAST_FLIGHT_DATE = 'test'
    NOW_ValidFare = -1
    Yesterday_fare = -1
    most_low_Fare_temp = 100000
    most_low_boolean = False
    most_high_Fare_temp = 0
    ValidFare_minus_most_low = 0
    ValidFare_minus_yesterday = 0
    sum_Fare = 0
    data_index = 0
    LAST_REMAIN_DAY = 45
    
    for i in range(len(data_array)):
        
        NOW_FLIGHT_DATE  = data_array[i,1]
        NOW_REMAIN_DAY   = data_array[i,0]
        NOW_ValidFare    = data_array[i,2]
        if(NOW_FLIGHT_DATE != LAST_FLIGHT_DATE):#如果得到的这个数据行的航班改变了，那就重新开始赋值并continue
            most_low_Fare_temp = 100000
            most_low_boolean = False
            most_high_Fare_temp = NOW_ValidFare
            ValidFare_minus_most_low = 0
            ValidFare_minus_yesterday = 0
            sum_Fare = NOW_ValidFare
            data_index = 1
            Yesterday_fare   = NOW_ValidFare
            LAST_FLIGHT_DATE = NOW_FLIGHT_DATE
            
            data_array[i,columns_len-6] = most_low_boolean
            data_array[i,columns_len-5] = ValidFare_minus_most_low
            data_array[i,columns_len-4] = ValidFare_minus_yesterday
            data_array[i,columns_len-3] = most_high_Fare_temp
            data_array[i,columns_len-2] = NOW_ValidFare
            data_array[i,columns_len-1] = most_low_Fare_temp
            continue
        else:
            if(NOW_ValidFare<=most_low_Fare_temp):
                most_low_Fare_temp = NOW_ValidFare
                most_low_boolean = True
            else:
                most_low_boolean = False
            if(NOW_ValidFare>=most_high_Fare_temp):
                most_high_Fare_temp = NOW_ValidFare
            ValidFare_minus_most_low  = NOW_ValidFare - most_low_Fare_temp
            if((LAST_REMAIN_DAY - NOW_REMAIN_DAY) ==1):
                ValidFare_minus_yesterday = NOW_ValidFare - Yesterday_fare
            else:
                ValidFare_minus_yesterday = NOW_ValidFare - Yesterday_fare
            data_array[i,columns_len-6] = most_low_boolean
            data_array[i,columns_len-5] = ValidFare_minus_most_low
            data_array[i,columns_len-4] = ValidFare_minus_yesterday
            data_array[i,columns_len-3] = most_high_Fare_temp
            data_array[i,columns_len-2] = sum_Fare/data_index
            data_array[i,columns_len-1] = most_low_Fare_temp
            #以往均价不包含当前价格， 以往的最低价格、以往的最高价格包含当前价格,这个可以试试别的
            sum_Fare += NOW_ValidFare
            data_index += 1
                
            Yesterday_fare   = NOW_ValidFare
            LAST_FLIGHT_DATE = NOW_FLIGHT_DATE

    return pd.DataFrame(data = data_array,columns = dateset_columns)


```

```python
#提取时间序列匹配特征

def create_dict(data,dict_temp):  
    flight_date_dict2 = {}
    for key in data.FLIGHT_DATE.drop_duplicates():
        list_temp = []
#         missing_flag = False
        for remain_day in range(45,0,-1):
            SEARCH_DATE = str(key)+':'+str(remain_day)
            if(SEARCH_DATE in dict_temp):
#                 debug_temp+=1
                price = dict_temp[SEARCH_DATE]
                list_temp.append(price['ValidFare'])
            else:
                list_temp.append('-1')
        flight_date_dict2[str(key)] = tuple(list_temp)
    return flight_date_dict2


def get_now_list(data_dict,FLIGHT_DATE,REMAIN_DAY):
    temp = int(45-REMAIN_DAY+1)
    return data_dict[FLIGHT_DATE][:temp]

def get_nearest_list(data_dict,FLIGHT_DATE,REMAIN_DAY,now_list):
    global ggp1
    global ggp2
    date1 = datetime.date(int(str(FLIGHT_DATE)[0:4]),int(str(FLIGHT_DATE)[5:7]),int(str(FLIGHT_DATE)[8:10]))
    delta=datetime.timedelta(days=-int(REMAIN_DAY))
    TICKET_DATE = date1 + delta
                          
    MAX_REMAIN_DAY = 45
    NEAREST_LENGTH = 18
    
    def get_wmae(list1,list2,remainday,weight_list):
        wMae = 0 
        miss_index = 1
        for i in range(MAX_REMAIN_DAY-remainday+1): 
            if((list1[int(i)]!='-1')&(list2[int(i)]!='-1')):
                wMae += abs(list1[i]-list2[i])*weight_list[i]
            elif(((list1[int(i)]=='-1')&(list2[int(i)]=='-1'))):
                wMae +=50*weight_list[i]*miss_index
                miss_index +=2
            else:
                wMae +=100*weight_list[i]*miss_index
                miss_index +=2
        return wMae/sum(weight_list)
    
    MAX_WEIGHT = 1
    if(REMAIN_DAY>15):#最佳参数是3，16
        MAX_WEIGHT = 3
    else:
        MAX_WEIGHT = 16
        
#     print(MAX_WEIGHT)
    weight_list = np.linspace(1, MAX_WEIGHT,MAX_REMAIN_DAY)
    nearest_dict = {}
#     FLAG = True
    for key,value in data_dict.items():
        #在这一步选择获选的已经起飞的航班时间序列作为待选时间序列
        if(((TICKET_DATE - datetime.date(int(str(key)[0:4]),int(str(key)[5:7]),int(str(key)[8:10]))).days>=0)&\
           ((TICKET_DATE - datetime.date(int(str(key)[0:4]),int(str(key)[5:7]),int(str(key)[8:10]))).days<=60)):
            nearest_dict[key] = value
    time1 = time.time()
 
    nearest_dict = sorted(nearest_dict.items(),
                          key= lambda x:get_wmae(x[1],now_list,REMAIN_DAY,weight_list),reverse = False)
    time2 = time.time()
#     ggp1+=(time2-time1)
    result_list = []
    if(len(nearest_dict)>NEAREST_LENGTH):
        for i in range(NEAREST_LENGTH):
            if(nearest_dict[i][1][45-REMAIN_DAY]!='-1'):#防止匹配到的时间序列在这个remainday下没有数据
                result_list.append(nearest_dict[i][0])
            else:
                continue
    else:
        return []
        
    return result_list

#给出前k个近邻时间序列在该REMAINDAY下的特征（区域最低价、
# 还有所有低于当前价格的低价的中位数、平均数，低于当前价格的天数占所有剩余天数的百分比，所有剩余天数的价格的均价）的均值
def extract_k_nearest_feature(dataset_list_dict,dataset_add_label_dict,FLIGHT_DATE,REMAIN_DAY):
    now_list    = get_now_list(dataset_list_dict,str(FLIGHT_DATE),REMAIN_DAY)
#     min_list = []
    result_temp = get_nearest_list(dataset_list_dict,FLIGHT_DATE,REMAIN_DAY,now_list)
    nearest_date_list=[]
    
    low_predict_temp = 0
    median_price_lower_than_now_price_temp = 0
    mean_price_lower_than_now_price_temp = 0
    lower_day_percent_temp = 0
    avg_price_temp = 0
    data_index = 0
    if(result_temp !=[]):
        for nearest_date in result_temp:
            key = nearest_date+':'+str(REMAIN_DAY)
            if(key in dataset_add_label_dict):
                if(dataset_add_label_dict[key]['low_predict']<10000):
                    low_predict_temp +=dataset_add_label_dict[key]['low_predict']
                    median_price_lower_than_now_price_temp +=dataset_add_label_dict[key]['median_price_lower_than_now_price']
                    mean_price_lower_than_now_price_temp += dataset_add_label_dict[key]['mean_price_lower_than_now_price']
                    lower_day_percent_temp += dataset_add_label_dict[key]['lower_day_percent']
                    avg_price_temp +=  dataset_add_label_dict[key]['avg_price']
                    data_index +=1
                else:#如果匹配到的航班的remainday下的未来最低价为10000，那说明这个数据的没有小于这个remainday下的数据，就用当天价格代替
                    low_predict_temp = now_list[-1]
                    median_price_lower_than_now_price_temp =   now_list[-1]
                    mean_price_lower_than_now_price_temp = now_list[-1]
                    lower_day_percent_temp = 0
                    avg_price_temp  = now_list[-1]
                    data_index = 1
    else:
        low_predict_temp = now_list[-1]
        median_price_lower_than_now_price_temp =   now_list[-1]
        mean_price_lower_than_now_price_temp = now_list[-1]
        lower_day_percent_temp = 0
        avg_price_temp  = now_list[-1]
        data_index = 1
        
    return low_predict_temp/data_index,median_price_lower_than_now_price_temp/data_index,\
        mean_price_lower_than_now_price_temp/data_index,lower_day_percent_temp/data_index,\
        avg_price_temp/data_index
        
def extract_feature2_2(data,dataset_add_label_dict):
    data_result = copy.deepcopy(data)
    dataset_list_dict = create_dict(data_result,dataset_add_label_dict) 
    data_result['k_nearest_low_predict_mean'] = -1
    data_result['k_nearest_median_price_lower_than_now_price_mean'] = -1
    data_result['k_nearest_mean_price_lower_than_now_price_mean'] = -1
    data_result['k_nearest_lower_day_percent_mean'] = -1
    data_result['k_nearest_avg_price_mean'] = -1
#     global ggp1
    
    column_lens = len(data_result.columns)
    
    data_array = data_result.as_matrix()
#     print(data_result.head())
    dataset_columns = data_result.columns
    p1 = 0
    for index_temp in range(len(data_array)):
#       temp = data_result.iloc[index_temp,]
        FLIGHT_DATE = data_array[index_temp][1]
        REMAIN_DAY =  data_array[index_temp][0]
        ValidFare = data_array[index_temp][2]
#         print('ValidFare=',str(ValidFare))
        pp1 = time.time()
        k_nearest_low_predict_mean,k_nearest_median_price_lower_than_now_price_mean,k_nearest_mean_price_lower_than_now_price_mean,\
            k_nearest_lower_day_percent_mean,k_nearest_avg_price_mean = \
            extract_k_nearest_feature(dataset_list_dict,dataset_add_label_dict,FLIGHT_DATE,REMAIN_DAY)
        pp2 = time.time()
        p1+=(pp2-pp1)
        data_array[index_temp,column_lens-5] = k_nearest_low_predict_mean
        data_array[index_temp,column_lens-4] = k_nearest_median_price_lower_than_now_price_mean
        data_array[index_temp,column_lens-3] = k_nearest_mean_price_lower_than_now_price_mean
        data_array[index_temp,column_lens-2] = k_nearest_lower_day_percent_mean
        data_array[index_temp,column_lens-1] = k_nearest_avg_price_mean
#     data_result = pd.DataFrame(data = data_array,columns = dataset_columns)
#     print('p1=',p1)
#     print('sort_time=',ggp1)
    return pd.DataFrame(data = data_array,columns = dataset_columns)

```

```python

def extract_feaature3_1(data):

    data['FLIGHT_DATE2'] = data.FLIGHT_DATE.apply(lambda x:x.replace('-',''))
    data['year'] = data.FLIGHT_DATE2.astype('str').apply(lambda x:x[0:4])
    data['month'] = data.FLIGHT_DATE2.astype('str').apply(lambda x:x[4:6])
    data['day'] = data.FLIGHT_DATE2.astype('str').apply(lambda x:x[6:8])
    data['flightdate_day_of_week'] = data.FLIGHT_DATE2.astype('str').apply(lambda x:date(int(x[0:4]),
                                                                      int(x[4:6]),int(x[6:8])).weekday()+1)

    data['in_0_10']  =  data['REMAIN_DAY']<=10
    data['in_10_20'] = (data['REMAIN_DAY']<=20)&(data['REMAIN_DAY']>10)
    data['in_20_30'] = (data['REMAIN_DAY']<=30)&(data['REMAIN_DAY']>20)
    data['in_30_40'] = (data['REMAIN_DAY']<=40)&(data['REMAIN_DAY']>30)
    data['in_40_50'] = (data['REMAIN_DAY']<=50)&(data['REMAIN_DAY']>40)

    data['TICKET_DATE2'] = data.TICKET_DATE.apply(lambda x:x.replace('-',''))
    data['ticketday_year']  = data.TICKET_DATE2.astype('str').apply(lambda x:x[0:4])
    data['ticketday_month'] = data.TICKET_DATE2.astype('str').apply(lambda x:x[4:6])
    data['ticketday_day']   = data.TICKET_DATE2.astype('str').apply(lambda x:x[6:8])
    data['ticketday_day_of_week'] = data.TICKET_DATE2.astype('str').apply(lambda x:date(int(x[0:4]),
                                                int(x[4:6]),int(x[6:8])).weekday()+1)
    return data
```
