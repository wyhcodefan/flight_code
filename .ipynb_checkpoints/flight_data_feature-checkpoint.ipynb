#从航空机票查询数据提取特征如下：
1.查询票价日期对应的年、月、日以及是一周的第几天；
2.航班起飞日期对应的年、月、日以及是一周的第几天；
3.查询票价日期与航班起飞日期相差的天数；
4.查询票价日期与航班起飞日期区间内最低票价（label）；
5.查询票价日期与航班起飞日期区间内平均最低票价；
6.上月平均最低票价
7.当前月所在季度平均最低票价

```python
import pandas as pd
import numpy as np
import operator
```

```python
fd = pd.read_csv('valid_data.csv')   #读取清洗好的数据，没有缺失值，remain_day<=60的数据
print(fd.columns)
print(fd.head(10))
```

```python
#确定训练集和测试集的日期区间
fd['ticket_date'] = pd.to_datetime(fd['ticket_date'])
fd['flight_date'] = pd.to_datetime(fd['flight_date'])
train_set = fd.loc[fd['ticket_date'] <= pd.datetime(2018,6,30)]
test_set = fd.loc[pd.datetime(2018,6,30) < fd['ticket_date']]
```

```python
#获取训练集里ticket_date对应的年月日
train_set['ticket_year'] = pd.to_datetime(train_set['ticket_date']).dt.year               # f1:年
train_set['ticket_month'] = pd.to_datetime(train_set['ticket_date']).dt.month             # f1:月
train_set['ticket_day'] = pd.to_datetime(train_set['ticket_date']).dt.day                 # f1:日
train_set['ticket_weekday'] = pd.to_datetime(train_set['ticket_date']).dt.weekday + 1       # f1:星期,+1的原因是星期一默认为0
train_set['flight_year'] = pd.to_datetime(train_set['flight_date']).dt.year               # f2:年
train_set['flight_month'] = pd.to_datetime(train_set['flight_date']).dt.month             # f2:月
train_set['flight_day'] = pd.to_datetime(train_set['flight_date']).dt.day                 # f2:日
train_set['flight_weekday'] = pd.to_datetime(train_set['flight_date']).dt.weekday + 1       # f2:星期  
train_set

#获取测试集里ticket_date对应的年月日
test_set['ticket_year'] = pd.to_datetime(test_set['ticket_date']).dt.year               # f1:年
test_set['ticket_month'] = pd.to_datetime(test_set['ticket_date']).dt.month             # f1:月
test_set['ticket_day'] = pd.to_datetime(test_set['ticket_date']).dt.day                 # f1:日
test_set['ticket_weekday'] = pd.to_datetime(test_set['ticket_date']).dt.weekday + 1       # f1:星期,+1的原因是星期一默认为0
test_set['flight_year'] = pd.to_datetime(test_set['flight_date']).dt.year               # f2:年
test_set['flight_month'] = pd.to_datetime(test_set['flight_date']).dt.month             # f2:月
test_set['flight_day'] = pd.to_datetime(test_set['flight_date']).dt.day                 # f2:日
test_set['flight_weekday'] = pd.to_datetime(test_set['flight_date']).dt.weekday + 1       # f2:星期  
test_set
```

```python
#将remain_day划分六个区间：0-10 == 1，11-20 == 2, 21-30 == 3, 31-40 == 4, 41-50 == 5, 51-60 == 6
import math
def remain_division(x):
    return math.ceil(x/10)

train_set['remain_day_interval'] = train_set['remain_day'].apply(remain_division)        #f3:remain_day，f4:remain_day所在区间
test_set['remain_day_interval'] = test_set['remain_day'].apply(remain_division)        #f3:remain_day，f4:remain_day所在区间
```

```python
train_set.to_csv('raw_train_data.csv',index = False)
test_set.to_csv('raw_test_data.csv',index = False)
```

```python
train_set = pd.read_csv('raw_test_data.csv')
#提取训练集航班数据有多少个航班期
flight_day_train = set(train_set.flight_date)


#计算各航班期不同remain_day的最低票价
frames = []
for i in flight_day_train:
    data_list = []
    for j in range(len(train_set)):
        if train_set.loc[j,'flight_date'] == i:
            data_list.append(j)
    temp_date_data = train_set.loc[data_list]
    temp_date_data.reset_index(drop = True, inplace = True)
    order_temp_date_data = temp_date_data.sort_values(['remain_day'],ascending = True)
    order_temp_date_data.reset_index(drop = True, inplace = True)
    order_temp_date_data['lowest_price'] = -1
    order_temp_date_data['ave_price'] = -1
    for k in range(0,len(order_temp_date_data)):   
        #提取各remain_day到航班日的最低价格
        order_temp_date_data.loc[k,'lowest_price'] = min(order_temp_date_data.loc[:k,'valid_price'])  #label
        #提取remain_day_max到当前remain_day的低价均值 <--f5
        order_temp_date_data.loc[len(order_temp_date_data)-1-k,'ave_price'] = order_temp_date_data.loc[len(order_temp_date_data)-1-k:len(order_temp_date_data)-1-0,'valid_price'].mean()
    frames.append(order_temp_date_data)
result_pool_train = pd.concat(frames)
result_pool_train.reset_index(drop = True, inplace = True)
#result_pool.columns = ['ticket_date','flight_date','remain_day','valid_price','ticket_year','ticket_month','ticket_day','ticket_weekday',\
#                      'flight_year','flight_month','flight_day','flight_weekday','lowest_price','ave_price']
print(result_pool_train.shape)
print(result_pool_train.head(10))
```

```python
result_pool_train.to_csv('result_pool_test.csv',index = False)
```

```python
#获取每月各remain_day区间的均价
result_pool = pd.read_csv('result_pool_test.csv')
ave_price_month = result_pool.groupby(['flight_month','remain_day']).agg('mean')
ave_price_month = pd.DataFrame(ave_price_month)

ave_price_month.reset_index(inplace = True)

ave_price_month = ave_price_month[['flight_month','remain_day','valid_price','lowest_price','ave_price']]


result_pool['ave_valid_price'] = -1                                #f6:同月同remain_day的均价
result_pool['ave_price_monday'] = -1                               #f7:同月remain_day_max到当前remain_day的均价的平均值
#result_pool['ave_price_lastmon'] = -1                              #f8:上月remain_day_max到当前remain_day的均价
for i in range(len(result_pool)):
    mon_id = result_pool.loc[i,'flight_month']
    remain_day_id = result_pool.loc[i,'remain_day']
    index_id = ave_price_month[(ave_price_month.flight_month == mon_id)&(ave_price_month.remain_day == remain_day_id)].index.tolist()
    index_id = index_id[0]

    result_pool.loc[i,'ave_valid_price'] = ave_price_month.loc[index_id,'valid_price']    #获取订票数据日期所在月的平均票价
    result_pool.loc[i,'ave_price_monday'] = ave_price_month.loc[index_id,'ave_price']     #获取订单数据日期所在月的最低票价
    #if mon_id == 1:
    #if mon_id == 7:
        #index_id = ave_price_month[(ave_price_month.flight_month == 12)&(ave_price_month.remain_day == remain_day_id)].index.tolist()
        #index_id = ave_price_month[(ave_price_month.flight_month == 12)&(ave_price_month.remain_day == remain_day_id)].index.tolist()
        #index_id = index_id[0]
        #result_pool.loc[i,'ave_price_lastmon'] = ave_price_month.loc[index_id,'ave_price']    #获取订票数据日期所在月的上月平均最低票价
        #result_pool.loc[i,'ave_price_lastmon'] = result_pool.loc[i,'ave_price_monday']  
    #else:
        #index_id = ave_price_month[(ave_price_month.flight_month == (mon_id - 1))&(ave_price_month.remain_day == remain_day_id)].index.tolist()
        #index_id = index_id[0]
        #result_pool.loc[i,'ave_price_lastmon'] = ave_price_month.loc[index_id,'ave_price']    #获取订票数据日期所在月的上月平均最低票价
```
